# eclat

```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing dataset and data preprocessing

datasheet = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transaction = []
for i in range(0,7501):
    transaction.append([str(datasheet.values[i,j]) for j in range(0,20)])
```

```python
# training the eclat model on dataset

from apyori import apriori
rules = apriori(transactions = transaction, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 3)
```

```python
# visualizing the results

# Displaying the first results coming directly from the output of the apriori function

results = list(rules)
results
```

```python
# creating a dataframe visualization

def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['product 1', 'Product 2', 'Support'])
```
